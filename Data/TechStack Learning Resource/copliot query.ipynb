```python
class LearningResource:
    def __init__(self):
        self.query_dict_list = {}
        self.queue = []
        self.driver_count = 9
        self.driver_list = []
        self.InitQueryList()
        self.InitAllDriver()
        
    def InitQueryList(self):   
        df = pd.read_csv("skills.csv", usecols=['Search Keyword', 'Resource Path'])
        for index, row in df.iterrows():
            key =  str(row['Search Keyword'])
            value = str(row['Resource Path'])
            if key is not self.query_dict_list:
                self.query_dict_list[key]= value
                self.queue.append(key)
        df = pd.read_csv("groups.csv", usecols=['Search Keyword', 'Resource Path'])
        for index, row in df.iterrows():
            key =  str(row['Search Keyword'])
            value = str(row['Resource Path'])
            if key is not self.query_dict_list:
                self.query_dict_list[key]= value
                self.queue.append(key)

    def InitDriverThread(self, index):
        self.driver_list.append(Copliot())
        print(f"Driver {index} initialized.")

    def InitAllDriver(self):
        threads = []
        for i in range(self.driver_count):
            t = threading.Thread(target=self.InitDriverThread, args=(i+1,))
            threads.append(t)
            t.start()

        for t in threads:
            t.join()
        print("Done")

    def LoginDriverThread(self, i): 
        no = i + 1
        id = f"copliot{no}@hotmail.com"
        pw = "NMT[901f]"
        result =  self.driver_list[i].Login(id, pw)
        if not result:
            print(f"driver {no} login failed.\n")
            
    def LoginDriver(self):
        threads = []
        for i in range(len(self.driver_list)):
            t = threading.Thread(target=self.LoginDriverThread, args=(i,))
            threads.append(t)
            t.start()
        for t in threads:
            t.join()

    def CloseDriverThread(self, driver):
        driver.CloseDriver()

    def CloseAllDriver(self):
        threads = []
        for driver in self.driver_list:
            t = threading.Thread(target=self.CloseDriverThread, args=(driver,))
            threads.append(t)
            t.start()

        for t in threads:
            t.join()
```
